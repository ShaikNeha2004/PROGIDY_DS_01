```python
import pandas as pd
import matplotlib.pyplot as plt
population_data = pd.read_csv('Population(1).csv', skiprows=4)
countries = ['United States', 'India', 'China', 'Brazil', 'Nigeria', 'Germany', 'Japan']
year = '2020'
population = population_data[population_data['Country Name'].isin(countries)]
plt.figure(figsize=(10, 6))
plt.bar(population['Country Name'], population[year], color='VIOLET')
plt.xlabel('Country')
plt.ylabel(f'Population in {year}')
plt.title(f'Population of Selected Countries in {year}')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
